```python
import pandas as pd
```

```python
data = [[1, '2015-01-01', 10], [2, '2015-01-02', 25], [3, '2015-01-03', 20], [4, '2015-01-04', 30]]
weather = pd.DataFrame(data, columns=['id', 'recordDate', 'temperature']).astype({'id':'Int64', 'recordDate':'datetime64[ns]', 'temperature':'Int64'})
```

```python
weather
```

```python
df = weather.sort_values(by='recordDate')
```

```python
df
```

```python
df['prev_rocordDate'] = df['recordDate'].shift()
df['prev_temprature'] = df['temperature'].shift()
```

```python
df
```

```python
pd.DataFrame(df[((df['recordDate'] - df['prev_rocordDate']).dt.days == 1) & (df['temperature'] > df['prev_temprature'])]['id']) # also can use datetime.timedelta(days=1)
```
