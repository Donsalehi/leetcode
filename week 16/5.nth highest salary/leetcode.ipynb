```python
import pandas as pd
```

```python
data = [[1, 100], [2, 200], [3, 300], [2, 300]]
employee = pd.DataFrame(data, columns=['Id', 'Salary']).astype({'Id':'Int64', 'Salary':'Int64'})
```

```python
employee
```

```python
def nth_highest_salary(employee: pd.DataFrame, N: int):
  unique_salaries = employee['Salary'].drop_duplicates().sort_values(ascending=False).reset_index(drop=True)
  if N >= 1 and len(unique_salaries) >= N:
    nth_max = unique_salaries.iloc[N-1]
  else:
    nth_max = None
  return pd.DataFrame({f'getNthHighestSalary({N})': [nth_max]})
```

```python
N = 1
nth_highest_salary(employee, N)
```

```python
N = 2
nth_highest_salary(employee, N)
```
