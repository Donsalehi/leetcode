```python
import pandas as pd
```

```python
data = [[1, 'Joe', 70000, 1], [2, 'Jim', 90000, 1], [3, 'Henry', 80000, 2], [4, 'Sam', 60000, 2], [5, 'Max', 90000, 1]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```

```python
employee
```

```python
department
```

```python
df = pd.merge(employee, department, left_on='departmentId', right_on='id', how='inner')
```

```python
df
```

```python
df.groupby(by="departmentId").apply(lambda x: x[x['salary'] == x['salary'].max()]).reset_index(drop=True)[['name_y', 'name_x', 'salary']].rename(columns={'name_y': 'Department', 'name_x': 'Employee', 'salary': 'Salary'})
```

###method 2

```python
data = [[1, 'Joe', 70000, 1], [2, 'Jim', 90000, 1], [3, 'Henry', 80000, 2], [4, 'Sam', 60000, 2], [5, 'Max', 90000, 1]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
```

```python
department = department.set_index('id')
```

```python
department
```

```python
employee = employee[employee.groupby(by='departmentId')['salary'].rank(method='min', ascending=False) == 1]
```

```python
employee
```

```python
employee['Department'] = employee['departmentId'].map(department['name']) # or use apply(lambda x: department.loc[x])
```

```python
employee
```

```python
employee.rename(columns={'name': 'Employee', 'salary': 'Salary'})[['Department', 'Employee', 'Salary']]
```
