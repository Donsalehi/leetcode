```python
import pandas as pd
```

```python
data = [[1, 'Alice'], [2, 'Bob'], [13, 'John'], [6, 'Alex'], [9, None]]
students = pd.DataFrame(data, columns=['student_id', 'student_name']).astype({'student_id':'Int64', 'student_name':'object'})
data = [['Math'], ['Physics'], ['Programming']]
subjects = pd.DataFrame(data, columns=['subject_name']).astype({'subject_name':'object'})
data = [[1, 'Math'], [1, 'Physics'], [1, 'Programming'], [2, 'Programming'], [1, 'Physics'], [1, 'Math'], [13, 'Math'], [13, 'Programming'], [13, 'Physics'], [2, 'Math'], [1, 'Math']]
examinations = pd.DataFrame(data, columns=['student_id', 'subject_name']).astype({'student_id':'Int64', 'subject_name':'object'})
```

```python
students
```

```python
subjects
```

```python
examinations
```

```python
combination = pd.merge(students, subjects, how="cross")
```

```python
combination
```

```python
attendance_count = examinations.groupby(["student_id", "subject_name"]).size().reset_index(name="attended_exams")
```

```python
attendance_count
```

```python
result = pd.merge(combination, attendance_count, how="left", on=["student_id", "subject_name"]).sort_values(by=["student_id", "subject_name"])
```

```python
result
```

```python
result['attended_exams'] = result['attended_exams'].fillna(0).astype(int)
```
