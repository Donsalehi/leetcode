```python
import pandas as pd
```

```python
data = [[4, 2], [5, 1], [6, 2], [1, 1], [2, 1], [3, 1], [7, 2]]
logs = pd.DataFrame(data, columns=['id', 'num']).astype({'id':'Int64', 'num':'Int64'})
```

```python
logs
```

```python
result_data = []

for num in logs['num'].unique():
  group_ids = sorted(logs[logs['num'] == num]['id'].tolist())

  if len(group_ids) >= 3:
    is_arithmetic = False
    diff = group_ids[1] - group_ids[0]
    for i in range(2, len(group_ids)):
      if group_ids[i] - group_ids[i-1] == diff:
        is_arithmetic = True
        break

    if is_arithmetic:
      result_data.append({'ConsecutiveNums ':num})
```

```python
pd.DataFrame(result_data)
```

```python
df = logs.sort_values(by='id',ascending=True)

num_list = list(df['num'])
id_list = list(df['id'])

consecutive_number_list = []

for i in range(len(num_list) - 2):
  if num_list[i] == num_list[i+1] == num_list[i+2] and id_list[i] + 1 == id_list[i+1] and id_list[i] + 2 == id_list[i+2]:
    consecutive_number_list.append(num_list[i])

result_df = pd.DataFrame({'ConsecutiveNums':consecutive_number_list})
result_df = pd.DataFrame(result_df['ConsecutiveNums'].unique(), columns=['ConsecutiveNums'])
result_df
```
