```python
import pandas as pd
```

```python
data = [[1, '2019-02-17', '2019-02-28', 5], [1, '2019-03-01', '2019-03-22', 20], [2, '2019-02-01', '2019-02-20', 15], [2, '2019-02-21', '2019-03-31', 30], [3, '2019-02-21', '2019-03-31', 30]]
prices = pd.DataFrame(data, columns=['product_id', 'start_date', 'end_date', 'price']).astype({'product_id':'Int64', 'start_date':'datetime64[ns]', 'end_date':'datetime64[ns]', 'price':'Int64'})
data = [[1, '2019-02-25', 100], [1, '2019-03-01', 15], [2, '2019-02-10', 200], [2, '2019-03-22', 30]]
units_sold = pd.DataFrame(data, columns=['product_id', 'purchase_date', 'units']).astype({'product_id':'Int64', 'purchase_date':'datetime64[ns]', 'units':'Int64'})
```

```python
prices
```

```python
units_sold
```

```python
prices['start_date'] = pd.to_datetime(prices['start_date'])
prices['end_date'] = pd.to_datetime(prices['end_date'])
prices
```

```python
units_sold['purchase_date'] = pd.to_datetime(units_sold['purchase_date'])
units_sold
```

```python
df = pd.merge(prices, units_sold, on='product_id')
```

```python
df
```

```python
df = df[(df['purchase_date'] >= df['start_date']) & (df['purchase_date'] <= df['end_date'])]
```

```python
df
```

```python
df['total_price'] = df['price'] * df['units']
```

```python
df
```

```python
df = df.groupby('product_id', as_index=False).agg(total_unit=('units', 'sum'),
                                                  total_price=('total_price', 'sum'))
```

```python
df
```

```python
df['average_price'] = (df['total_price'] / df['total_unit']).round(2).fillna(0)
```
